**Role of Gender in Customer Churn**

```python
#Total Customers
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(3,3))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Total Customers by Gender')
plt.show()
```

```python
#Exited Customers
gender_counts = df[df['Exited'] == 1]['Gender'].value_counts()

plt.figure(figsize=(3,3))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Exited Customers by Gender')
plt.show()
```

```python
total_count_per_gender = df['Gender'].value_counts()

# Calculate the count of exits for each gender
exited_count_per_gender = df[df['Exited'] == 1]['Gender'].value_counts()

# Calculate the percentage of exits for each gender
percentage_exits_per_gender = (exited_count_per_gender / total_count_per_gender) * 100

plt.figure(figsize=(3,3))
percentage_exits_per_gender.plot(kind='bar', color='skyblue')
plt.title('Percentage of Exits by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage of Exits')
plt.xticks(rotation=0)  # Rotate x-labels if necessary
plt.tight_layout()
plt.show()
```

```python
total_count_per_country = df['Geography'].value_counts()

# Calculate the count of exits for each gender
exited_count_per_country = df[df['Exited'] == 1]['Geography'].value_counts()

# Calculate the percentage of exits for each gender
percentage_exits_per_country = (exited_count_per_country / total_count_per_country) * 100

plt.figure(figsize=(3,3))
percentage_exits_per_country.plot(kind='bar', color='skyblue')
plt.title('Percentage of Exits by Geography')
plt.xlabel('Country')
plt.ylabel('Percentage of Exits')
plt.xticks(rotation=0)  # Rotate x-labels if necessary
plt.tight_layout()
plt.show()
```

## Data Cleaning

Removing unnecessary columns

```python
clean_data = df.drop(['RowNumber', 'CustomerId','Surname'], axis=1)
```

Checking Null Values

```python
clean_data.isna().sum()
```

Label Encoding Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
clean_data['Geography'] = encoder.fit_transform(clean_data['Geography'])
clean_data['Gender'] = encoder.fit_transform(clean_data['Gender'])
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(clean_data.corr(),annot=True)
```

```python
clean_data.info()
```

```python
x = clean_data.drop('Exited', axis=1)
y = clean_data['Exited']
```

```python
x.head()
```

**train-test split**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

## ML Models

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

###Logistic Regression

```python
np.random.seed(42)
```

```python
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression(random_state=42)
logReg.fit(x_train,y_train)

print("Accuracy on training data", logReg.score(x_train, y_train))
print("Accuracy on testing data", logReg.score(x_test, y_test))
```

```python
cv_scores_log_reg = cross_val_score(logReg, x, y, cv=kf)
print(cv_scores_log_reg.mean())
```

```python
y_pred = logReg.predict(x_test)
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
cm_display.plot()
```

```python
sensitivity = cm[1,1]/(cm[1,1] + cm[1,0])
print("Sensitivity:", sensitivity)
```

```python
specificity = cm[0,0]/(cm[0,0] + cm[0,1])
print("Specificity:",specificity)
```

Logistic Regression gives extremely low recall i.e. it is not able to identify exited customers well.

###Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
np.random.seed(123)
dt = DecisionTreeClassifier(class_weight={0: 0.1, 1: 0.9}, random_state=123)
dt.fit(x_train, y_train)

print("Accuracy on training data", dt.score(x_train, y_train))
print("Accuracy on testing data", dt.score(x_test, y_test))
```

We use hyper-parameter tuning to remove overfitting

```python
print("Max depth of decision tree:", dt.get_depth())

from sklearn import tree
tree.plot_tree(dt,feature_names = x_test.columns);
```

```python
dt = DecisionTreeClassifier(max_depth=5)
dt.fit(x_train, y_train)

print("Accuracy on training data", dt.score(x_train, y_train))
print("Accuracy on testing data", dt.score(x_test, y_test))
```

```python
tree.plot_tree(dt,feature_names = x_test.columns);
```

```python
cv_scores_dt = cross_val_score(dt, x, y, cv=kf)
print(cv_scores_dt.mean().round(4))
```

```python
y_pred = dt.predict(x_test)
```

```python
cm = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
cm_display.plot()
```

```python
sensitivity = cm[1,1]/(cm[1,1] + cm[1,0])
print("Sensitivity:", sensitivity)
```

Decision tree gives better recall than Logistic Regression but is still unsatisfactory.

### Support Vector Machine

```python
#Support Vector Machine
from sklearn import svm
np.random.seed(129)
svm_model = svm.SVC(C=25, random_state = 129)
svm_model.fit(x_train, y_train)

print("Accuracy on testing data", svm_model.score(x_train, y_train))
print("Accuracy on training data", svm_model.score(x_test, y_test))
```

```python
cv_scores_svm = cross_val_score(svm_model, x, y, cv=kf)
print(cv_scores_svm.mean().round(4))
```

```python
y_pred = svm_model.predict(x_test)

cm = metrics.confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
cm_display.plot()
```

### k-nearest neighbors

```python
# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

print("Accuracy on testing data", knn.score(x_train, y_train))
print("Accuracy on training data", knn.score(x_test, y_test))
```

Increasing neighbors

```python
# K-Nearest Neighbors

knn2 = KNeighborsClassifier(n_neighbors=50)
knn2.fit(x_train, y_train)

print("Accuracy on training data", knn2.score(x_train, y_train))
print("Accuracy on testing data", knn2.score(x_test, y_test))
```

```python
cv_scores_knn = cross_val_score(knn2, x, y, cv=kf)
print(cv_scores_knn.mean().round(4))
```

```python
y_pred = knn2.predict(x_test)

from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
```

###Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
np.random.seed(49)
rf_model = RandomForestClassifier(n_estimators=500, random_state=49)
rf_model.fit(x_train, y_train)

print("Accuracy on training data", rf_model.score(x_train, y_train))
print("Accuracy on testing data", rf_model.score(x_test, y_test))
```

Hyperparameter tuning

```python
cv_scores_rf = cross_val_score(rf_model, x, y, cv=kf)
print(cv_scores_rf.mean().round(4))
```

## Conclusion
Since predicting exiting customers accurately is more important than predicting not exiting customers, we would select the model that provides greater recall and satisfactory accuracy. As Random Forest provides greater recall i.e. 0.74, it predicts customer churn more accurately than other models.
